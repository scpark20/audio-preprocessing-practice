```python
import numpy as np

import matplotlib.pyplot as plt
```

## 삼각함수 기초

#### $cos(n)$는 $2\pi\simeq6.2832$를 주기로 갖는 함수, $cos(0)=1$


```python
n = np.linspace(0, 31, 32) 
y = np.cos(n) 

plt.figure(figsize=[18, 2])
plt.plot(n, y, color = 'red', marker='o') 
plt.title("cos(n)") 
plt.xlabel("n") 
plt.ylabel("y") 
plt.show() 
```

#### 연습1. $sin(n)$함수를 그려보고, $sin(0)$값은 무엇인지 확인합시다.


### 크기(amplitude)를 변화시켜 봅시다.
#### $Acos(n)$는 크기가 $A$, $2\pi\simeq6.2832$를 주기로 갖는 함수, $cos(0)=A$


```python
A = 4
n = np.linspace(0, 31, 32) 
y = A * np.cos(n) 

plt.figure(figsize=[18, 2])
plt.plot(n, y, color = 'red', marker='o') 
plt.title("4cos[n]") 
plt.xlabel("n") 
plt.ylabel("y") 
plt.show() 
```

#### 연습2. $4sin[n]$함수를 그려보고, $4sin[0]$값은 무엇인지 확인합시다.


### 주기(period)를 변화시켜 봅시다.
#### $Acos[2\pi n/p], Asin[2\pi n/p]$는 크기가 A, 주기가 $p$인 함수


```python
n = np.linspace(0, 31, 32) 
A = 4
period = 8
y1 = A * np.cos(2 * np.pi * n / period) 
y2 = A * np.sin(2 * np.pi * n / period) 

plt.figure(figsize=[18, 2])
plt.plot(n, y1, color = 'red', marker = "o") 
plt.plot(n, y2, color = 'blue', marker = "o") 
plt.title("4cos(2πn/8), 4sin(2πn/8)") 
plt.xlabel("n") 
plt.ylabel("y") 
plt.show() 
```

#### 연습3. 주기가 16, 8, 4, 2, 1인 cos, sin함수를 차례대로 그려보고, 그려진 모양에 대해서 생각해봅시다.


### 구간 내의 진동 횟수를 정해봅시다.
#### $Acos[2\pi kn/N]$는 크기가 A이고, N구간 내에서 k번 진동하는 함수, 즉 $N/k$를 주기로 갖는 cos함수


```python
# 구간 길이 지정
N = 32
# 해당 구간 안에서 몇 번 진동이 반복 될지 지정
k = 4
n = np.linspace(0, N-1, N) 
y1 = np.cos(2 * np.pi * k * n / N) 
y2 = np.sin(2 * np.pi * k * n / N) 

plt.figure(figsize=[18, 2])
plt.plot(n, y1, color = 'red', marker = "o") 
plt.plot(n, y2, color = 'blue', marker = "o")
plt.title("cos((2π/p)x), sin((2π/p)x)")
plt.xlabel("x") 
plt.ylabel("y") 
plt.show() 
```

### 위상(phase)를 변화시켜봅시다.
#### $Acos[2\pi kn/N+\phi]$ 크기가 A이고, N구간 내에서 k번 진동하는 함수이며, 왼쪽으로 $\phi$만큼 이동한 함수
#### $Acos[2\pi kn/N+0.5\pi]=-Asin[2\pi kn/N]$, cos함수를 $0.5\pi$만큼 왼쪽으로 이동시키면, -sin함수가 됩니다.


```python
# 구간 길이 지정
N = 32
# 해당 구간 안에서 몇 번 진동이 반복 될지 지정
k = 4
n = np.linspace(0, N-1, N) 
y1 = np.cos(2 * np.pi * k * n / N + 0.5 * np.pi) 
y2 = -np.sin(2 * np.pi * k * n / N) + 0.1 #그래프가 안겹치게 0.1더함

plt.figure(figsize=[18, 2])
plt.plot(n, y1, color = 'red', marker = "o") 
plt.plot(n, y2, color = 'blue', marker = "o")
plt.title("cos((2π/p)x), -sin((2π/p)x)")
plt.xlabel("x") 
plt.ylabel("y") 
plt.show() 
```

#### 연습4. sin함수를 왼쪽으로 $0.5\pi$만큼 이동시켜 cos함수와 모양이 일치하도록 만들고 그래프를 그려 비교해 봅시다.


### 위상(phase)이 다른 여러 함수

```python
# 구간 길이 지정
N = 32
# 해당 구간 안에서 몇 번 진동이 반복 될지 지정
k = 4
n = np.linspace(0, N-1, N) 

plt.figure(figsize=[18, 2])
for phase in [0.0*np.pi, 0.1*np.pi, 0.2*np.pi, 0.3*np.pi, 0.4*np.pi]:
    y = np.cos(2 * np.pi * k * n / N + phase) 
    plt.plot(n, y, marker = "o", alpha=0.6) 
    
plt.xlabel("x") 
plt.ylabel("y") 
plt.show() 
```

#### 연습5. 위상(phase)값은 대개 $[-\pi, \pi)$ 또는 $[0, 2\pi)$ 범위 안의 값으로 표현합니다. 다양한 위상을 같는 삼각함수 그래프들을 그려보면서 그 이유를 생각해봅시다.

### 삼각함수 덧셈 공식
#### $sin(\alpha+\beta)=sin(\alpha)cos(\beta)+cos(\alpha)sin(\beta)$
#### $cos(\alpha+\beta)=cos(\alpha)cos(\beta)-sin(\alpha)sin(\beta)$

### 크기가 주어지고 위상값이 더해진 $cos$식은 위 공식에 의해 다음과 같이 계수(coefficient)가 곱해진 $cos$과 $sin$의 합으로 나타낼 수 있습니다.
#### $\color{blue}4cos[2\pi kn/N+\color{red}{0.1\pi}]=\color{blue}4(cos[2\pi kn/N]\color{red}{cos(0.1\pi)}\color{red}-sin[2\pi kn/N]\color{red}{sin(0.1\pi)})$


```python
#크기 지정
A = 4
# 구간 길이 지정
N = 32
# 해당 구간 안에서 몇 번 진동이 반복 될지 지정
k = 4
n = np.linspace(0, N-1, N) 
y1 = A * np.cos(2 * np.pi * k * n / N - 0.5 * np.pi + 0.1 * np.pi) 
y2 = A * (np.cos(2 * np.pi * k * n / N - 0.5 * np.pi) * np.cos(0.1 * np.pi) - \
     np.sin(2 * np.pi * k * n / N - 0.5 * np.pi) * np.sin(0.1 * np.pi)) + 0.3 #그래프가 안겹치게 0.3더함

plt.figure(figsize=[18, 2])
plt.plot(n, y1, color = 'red', marker = "o") 
plt.plot(n, y2, color = 'blue', marker = "o")
plt.xlabel("x") 
plt.ylabel("y") 
plt.show() 
```

#### Euler's formula
#### $e^{j\phi} = \cos \phi + j\sin \phi$

#### $cos[2\pi kn/N]$과 $sin[2\pi kn/N]$은 위상값이 더해진 신호$cos[2\pi kn/N+\phi]$이나 $sin[2\pi kn/N+\phi]$을 표현하기 위한 기저(basis)를 형성합니다. 편의상 sin성분을 허수축(imaginary axis)으로 두면 Euler's formula에 의해 훨씬 간단하게 표현할 수 있습니다.
